```python
import pandas as pd

def replace_multiple(df, column, replacements):
    # Verificar si la columna es de tipo cadena (string)
    if df[column].dtype == 'O':
        for old, new in replacements.items():
            df[column] = df[column].str.replace(old, new)
    else:
        print(f"La columna {column} no es de tipo cadena (string) y no se realizaron cambios.")

    return df[column]

# Ejemplo de uso:
# Supongamos que df es tu DataFrame original

# Definir los reemplazos deseados
replacements = {
    'Ã': 'Á',
    'Ã‰': 'É',
    'Ã': 'Í',
    'Ã“': 'Ó',
    'Ãš': 'Ú',
    'Ã‘': 'Ñ',
    'Ã¡': 'á',
    'Ã©': 'é',
    'Ã­': 'í',
    'Ã³': 'ó',
    'Ãº': 'ú',
    'Ã±': 'ñ'
}


```

```python
#hacer una lista de las columnas que tengan valores de tipo object

accidentes.select_dtypes(include=['object']).columns

# aplivar la función de reemplazo para las columnas que tengan valores de tipo object

for col in accidentes.select_dtypes(include=['object']).columns:
    accidentes[col] = replace_multiple(accidentes, col, replacements)

accidentes.head()

```

```python
# cambiar los nombres de las columnas "	ï»¿num_expediente" y "estado_meteorolÃ³gico" por "num_expediente" y "estado_meteorologico"

accidentes.rename(columns={'ï»¿num_expediente':'num_expediente', 'estado_meteorolÃ³gico':'estado_meteorologico'}, inplace=True)

accidentes.head()
```

```python
# hacer una funcion que convierta todos los strings a minusculas del dataframe

def minusculas(df):
    for i in df:
        if df[i].dtype == 'object':
            df[i] = df[i].str.lower()
    return df

minusculas(accidentes)

accidentes.head()
```

```python
def mapear_rango_edad(df):
    mapeo_edad = {
        'menor de 5 años': 'niño',
        'de 6 a 9 años': 'niño',
        'de 10 a 14 años': 'niño',
        'de 15 a 17 años': 'adolescente',
        'de 18 a 20 años': 'joven',
        'de 21 a 24 años': 'joven',
        'de 25 a 29 años': 'joven',
        'de 30 a 34 años': 'adulto',
        'de 35 a 39 años': 'adulto',
        'de 40 a 44 años': 'adulto',
        'de 45 a 49 años': 'adulto',
        'de 50 a 54 años': 'adulto',
        'de 55 a 59 años': 'adulto',
        'de 60 a 64 años': 'adulto',
        'de 65 a 69 años': 'mayor',
        'de 70 a 74 años': 'mayor',
        'más de 74 años': 'mayor'
    }

    df.rango_edad = df.rango_edad.replace(mapeo_edad)
    return df

# Llamada a la función
df = mapear_rango_edad(accidentes)

accidentes.rango_edad.value_counts()
```

```python
# rellenamos los missing values del dataframe con la moda de cada columna

accidentes.fillna(df.mode().iloc[0], inplace=True)

accidentes.head()
```

```python
# eliminamos los elemnetos de la columna de localizacion que esten a partir de la coma

accidentes['localizacion'] = accidentes['localizacion'].str.split(',').str[0]

accidentes.head()
```

```python
# eliminamos las columnas que tengan mas del 90% de valores nulos a traves de una funcion

def eliminar_nulos(df, porcentaje):
    for i in df:
        if df[i].isnull().sum() > (len(df) * porcentaje):
            df.drop(i, axis=1, inplace=True)
    return df

# llamada a la funcion

enero = eliminar_nulos(enero, 0.9)

enero.shape
```
