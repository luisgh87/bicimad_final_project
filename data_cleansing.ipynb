```python
import pandas as pd

def replace_multiple(df, column, replacements):
    # Verificar si la columna es de tipo cadena (string)
    if df[column].dtype == 'O':
        for old, new in replacements.items():
            df[column] = df[column].str.replace(old, new)
    else:
        print(f"La columna {column} no es de tipo cadena (string) y no se realizaron cambios.")

    return df[column]

# Ejemplo de uso:
# Supongamos que df es tu DataFrame original

# Definir los reemplazos deseados
replacements = {
    'Ã': 'Á',
    'Ã‰': 'É',
    'Ã': 'Í',
    'Ã“': 'Ó',
    'Ãš': 'Ú',
    'Ã‘': 'Ñ',
    'Ã¡': 'á',
    'Ã©': 'é',
    'Ã­': 'í',
    'Ã³': 'ó',
    'Ãº': 'ú',
    'Ã±': 'ñ'
}


```

```python
#hacer una lista de las columnas que tengan valores de tipo object

df.select_dtypes(include=['object']).columns

# aplivar la función de reemplazo para las columnas que tengan valores de tipo object

for col in df.select_dtypes(include=['object']).columns:
    df[col] = replace_multiple(df, col, replacements)

df.head()

```

```python
# hacer una funcion que convierta todos los strings a minusculas del dataframe

def minusculas(df):
    for i in df:
        if df[i].dtype == 'object':
            df[i] = df[i].str.lower()
    return df

minusculas(df)

df.head()
```
